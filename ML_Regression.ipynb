```python
def custom_hist(training_set, title,  xlabel, ylabel='ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾', bins=None):
    figsize = (20,6)
    plt.figure(figsize=figsize)
    plt.grid(True)
    plt.title(title)
    plt.hist(training_set, training_set.max().astype(training_set.dtype) if bins is None else bins)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
data = data.drop('car_ID', axis=1)
data = data.drop('CarName', axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
ohe = OneHotEncoder()

for i in data.columns:
    if data[i].dtype == 'object':
        transformed = ohe.fit_transform(data[[i]])
        data[ohe.categories_[0]] = transformed.toarray()
        data = data.drop(i, axis = 1)
data
```

```python
X = data.drop('price', axis = 1)
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import SGDRegressor, LinearRegression, Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
def compare_models(models, X_train, X_test, y_train, y_test):
    for model in models:
        model = model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        print(f"------------------------")
        print(f"MODEL: {str(model)}")
        print(f"MAPE:  {mean_absolute_percentage_error(y_test, y_pred)}")
        print(f"MAE:   {mean_absolute_error(y_test, y_pred)}")
        print(f"CVS:   {cross_val_score(model, X_train, y_train, cv=5, scoring = 'neg_mean_absolute_percentage_error')}")
        print(f"------------------------")
```

```python
models = [LinearRegression(), GradientBoostingRegressor(), SGDRegressor(), Ridge()]
```

```python
compare_models(models, X_train, X_test, y_train, y_test)
```

```python
model = GradientBoostingRegressor().fit(X_test, y_test)
y_pred = model.predict(X_test)
```

```python
print(mean_absolute_percentage_error(y_pred, y_test))
```

```python
gs = GridSearchCV(estimator=GradientBoostingRegressor(),param_grid={'loss': ('squared_error', 'absolute_error'), 'n_estimators': [100, 500, 1000], 'max_depth': [3,4,5]})
```

```python
gs.fit(X_train, y_train)
```
